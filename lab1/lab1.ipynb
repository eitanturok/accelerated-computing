# Question 1

*How does the run time of the scalar GPU implementation compare to the scalar CPU implementation? What factors do you think might contribute to each of their run times?*

The scalar CPU implementation runs in 40ms and the scalar GPU implementation runs in 314ms. GPUs are faster than CPUs for highly parallelize tasks, but here we do not parallelize anything which is why the GPU does not give a speedup.

*Commands:*
